```python
# evaluating two regression and classifier models
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import precision_score, accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/DSML/nigeria_houses.csv')
df.head()
```

```python
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

#create a labelencoder instance
le_title = LabelEncoder()
le_town = LabelEncoder()
le_state = LabelEncoder()
```

```python
#Apply label encoding to each categorical column

df['title'] = le_title.fit_transform(df['title'])
df['town'] = le_town.fit_transform(df['town'])
df['state'] = le_state.fit_transform(df['state'])
```

```python
df.head()
```

```python
x = df.drop(columns=['price'])
y = df['price']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
# invoking suport vector machine regressor

from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
svr_model = SVR(kernel='linear')
svr_model.fit(x_train, y_train)
```

```python
x_test
```

```python
svr_prediction = svr_model.predict(x_test)
svr_prediction
```

```python
len(svr_prediction)
```

```python
y_test
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
```

```python
logistic_prediction = logistic_model.predict(x_test)
logistic_prediction
```

```python
len(logistic_prediction)
```

```python
# define evaluation metrics
metrics = {
    "mean_squared_error": mean_squared_error,
    "mean_absolute_error": mean_absolute_error
}

results = {}
for name, func in metrics.items():
  # calculate scores for each model
  svr_score = func(y_test, svr_prediction)
  logistic_score = func(y_test, logistic_prediction)

  # store results in the dictionary
  results[name] = {
      "Support Vector Regression": svr_score,
      "Logistic Regression": logistic_score
  }
```

```python
results
```

```python
# Random Forest Model

random_forest_model = RandomForestClassifier()
random_forest_model.fit(x_train, y_train)
random_forest_prediction = random_forest_model.predict(x_test)
```

```python
random_forest_prediction
```

```python
y_test
```

```python
metrics = {
    "accuracy_score": accuracy_score,
    "precision_score": precision_score
}

results_random_forest = {}
for name, func in metrics.items():
  # calculate scores for each model
  random_forest_score = func(y_test, random_forest_prediction)

  # store results in the dictionary
  results_random_forest[name] = {
      "Random Forest Classifier": random_forest_score
  }
```

```python
results_random_forest
```

```python
decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(x_train, y_train)
decision_tree_prediction = decision_tree_model.predict(x_test)
```

```python
metrics = {
    "accuracy_score": accuracy_score,
    "precision_score": precision_score
}

results_decision_tree = {}
for name, func in metrics.items():
  # calculate scores for each model
  decision_tree_score = func(y_test, decision_tree_prediction)

  # store results in the dictionary
  results_decision_tree[name] = {
      "Decision Tree Classifier": decision_tree_score
  }
```

```python
results_decision_tree
```

```python
print('model evaluation results:\n')

# Evaluate Regression Models
print("Regression Model Evaluation:")
regression_metrics = {
    "mean_squared_error": mean_squared_error,
    "mean_absolute_error": mean_absolute_error
}
regression_results = {}
for name, func in regression_metrics.items():
  svr_score = func(y_test, svr_prediction)
  logistic_score = func(y_test, logistic_prediction)
  regression_results[name] = {
      "Support Vector Regression": svr_score,
      "Logistic Regression": logistic_score
  }
  print(f"{name}:\n Support Vector Regression Score: {svr_score:.2f}\n Logistic Regression Score: {logistic_score:.2f}\n")

# Evaluate Classification Models
print("\nClassification Model Evaluation:")
classification_metrics = {
    "accuracy_score": accuracy_score,
    "precision_score": precision_score
}
classification_results = {}
for name, func in classification_metrics.items():
  random_forest_score = func(y_test, random_forest_prediction)
  decision_tree_score = func(y_test, decision_tree_prediction)
  classification_results[name] = {
      "Decision Tree Classifier": decision_tree_score,
      "Random Forest Classifier": random_forest_score
  }
  print(f"{name}:\n Decision Tree Classifier Score: {decision_tree_score:.2f}\n Random Forest Classifier Score: {random_forest_score:.2f}\n")


# Print summary of best model per metric (for classification)
print("\nBest Classification Model per Metric:")
for metric, scores in classification_results.items():
    best_model = max(scores, key=scores.get)
    print(f"{metric}: {best_model}")

# Print summary of best model per metric (for regression)
print("\nBest Regression Model per Metric:")
for metric, scores in regression_results.items():
    # For regression, lower error is better
    best_model = min(scores, key=scores.get)
    print(f"{metric}: {best_model}")
```
